**1.What is the difference between a function and a method in Python?**

* **Function:**
  - It is a defined block of code that performs a specific task, which can be used repeatitively.

  - Method is called along with the name of it's class.

  ```

      sum = num1+num2
  ```    

```
  return x+y
```

````python
#7.  Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_line_by_line(filename):
  with open (filename,"r") as file:
    for line in file:
      yield line.strip()

file_content = """This is first line,
this is second line,
this is third line"""

with open("sample.txt", "w") as f:
  f.write(file_content)

for line in read_file_line_by_line("sample.txt"):
  print(line)






````

````python
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

def fahren(a):
  return a*9/5+32

celsius_day_temp = [32,34,36,38,36,35]
fahren_day_temp = list(map(fahren, celsius_day_temp))
print(fahren_day_temp)

````

````python
#10.  Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(text):
    return ''.join(filter(lambda x: x.lower() not in 'aeiou', text))


print(remove_vowels("Hello Python"))
````
